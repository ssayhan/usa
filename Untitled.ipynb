```python
income = pd.to_numeric(income, errors='coerce')
```

```python
result = df_21.merge(income, on='Location', left_index=True, right_index=True)
```

```python
result.head()
```

```python
result['White_Pop_Per'] = (result['White'] / result['Total'])*100
result['Black_Pop_Per'] = (result['Black'] / result['Total'])*100
result['Hispanic_Pop_Per'] = (result['Hispanic'] / result['Total'])*100
result['Asian_Pop_Per'] = (result['Asian'] / result['Total'])*100
result['American Indian/Alaska Native_Pop_Per'] = (result['American Indian/Alaska Native'] / result['Total'])*100
result['Native Hawaiian/Other Pacific Islander_Pop_Per'] = (result['Native Hawaiian/Other Pacific Islander'] / result['Total'])*100
result['Multiple_Races_Pop_Per'] = (result['Multiple Races'] / result['Total'])*100
```

```python
result.rename(columns={'Location':'State'}, inplace=True)
```
